```python
import pandas as pd
import pysam
pandora_multisample_vcf_ref = "/home/leandro/git/snippy_calls_gene_distance/data/pandora_multisample.vcf_ref.fa"
pandora_multisample_matrix_filepath="/home/leandro/git/snippy_calls_gene_distance/data/pandora_multisample.matrix"
```

```python
pandora_multisample_matrix = pd.read_csv(pandora_multisample_matrix_filepath, sep="\t")
max_number_of_samples = len(pandora_multisample_matrix.columns) - 1
nb_of_samples_gene_is_present = pandora_multisample_matrix.iloc[ : , 1:].agg(["sum"], axis=1)
pandora_multisample_matrix["nb_of_samples_gene_is_present"] = nb_of_samples_gene_is_present
genes_in_all_samples_df = pandora_multisample_matrix.query("nb_of_samples_gene_is_present == @max_number_of_samples")
genes_in_all_samples_list = genes_in_all_samples_df.iloc[:, 0].to_list()
genes_in_all_samples_list
```

```python
with pysam.FastxFile(pandora_multisample_vcf_ref) as fasta_file, open("genes_in_all_sample.fa", "w") as fout:
    for fasta_record in fasta_file:
        gene_name = fasta_record.name
        if gene_name in genes_in_all_samples_list:
            print(fasta_record, file=fout)
```

```python
import pandas as pd

variant_calls_probeset_report = pd.read_csv("/home/leandro/git/snippy_calls_gene_distance/data/hps/nobackup/research/zi/leandro/pandora1_paper/analysis_output_pandora_default_filters_snippy_all_bwa_seed_11/precision/reports_from_probe_mappings/063_STEC/all/snippy_NZ_MIPU01000001.1/coverage_filter_Not_App/strand_bias_filter_Not_App/gaps_filter_Not_App/variant_calls_probeset_report.tsv", sep="\t")
variant_calls_probeset_report = variant_calls_probeset_report.head()
display(variant_calls_probeset_report)
```

```python
def get_query_probes_info(variant_calls_probeset_report):
    query_probes_infos = []
    
    for index, row in variant_calls_probeset_report.iterrows():
        query_probe_header, classification = row["query_probe_header"], row["classification"]
        field_to_value_query_probe = {}
        query_probe_header_split = query_probe_header.split(";")
        for field_and_value in query_probe_header_split:
            field_and_value_split = field_and_value.split("=")
            if len(field_and_value_split) == 2:
                field, value = field_and_value_split
                field_to_value_query_probe[field] = value
        field_to_value_query_probe["POS"] = int(field_to_value_query_probe["POS"])
        field_to_value_query_probe["GT_CONF"] = float(field_to_value_query_probe["GT_CONF"])
        field_to_value_query_probe["classification"] = float(classification)
        query_probes_infos.append(field_to_value_query_probe)
    
    return query_probes_infos


query_probes_infos = get_query_probes_info(variant_calls_probeset_report)
query_probes_infos


```

```python
# >CHROM=NZ_CP007133.1;SAMPLE=063_STEC;POS=58;INTERVAL=[25,26);SVTYPE=snp;GT_CONF=3022.110107421875;
# >CHROM=(contig_ref_gene);SAMPLE=<ref_or_truth_id_truth_gene>;POS=58 - should be between start_ref_gene,stop_ref_gene

edit_distances_csv = pd.read_csv("/home/leandro/git/snippy_calls_gene_distance/evaluating_snippy_calls_output/edit_distances/GC00000153_1~~~063_STEC~~~NZ_MIPU01000001.1.edit_distance.csv")
display(edit_distances_csv)

contig_ref_gene = edit_distances_csv["contig_ref_gene"][0]
start_ref_gene = int(edit_distances_csv["start_ref_gene"][0])
stop_ref_gene = int(edit_distances_csv["stop_ref_gene"][0])
edit_distance = float(edit_distances_csv["edit_distance"][0])

edit_distances = []
for query_probes_info in query_probes_infos:
    if contig_ref_gene == query_probes_info['>CHROM'] and start_ref_gene <= query_probes_info['POS']*100000 and query_probes_info['POS'] <= stop_ref_gene:
        edit_distances.append(edit_distance)
    else:
        edit_distances.append(nan)

variant_calls_probeset_report["edit_distance"] = edit_distances
variant_calls_probeset_report
```
